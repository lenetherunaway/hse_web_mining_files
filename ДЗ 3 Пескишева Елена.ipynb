# Задача 1

```python
import requests
```

```python
from bs4 import BeautifulSoup
import urllib 
```

```python
def get_all_links_wiki(url):
    import requests
    response = requests.get(url)
    from bs4 import BeautifulSoup
    import urllib 
    soup = BeautifulSoup(response.text, 'html.parser')
    links = []
    for link in soup.find("div", {"id":"mw-content-text"}).findAll('a'):
        if 'wiki' in str(link.get('href')) and not('index' in str(link.get('href'))) and not ('svg' in str(link.get('href'))):
            links.append(link.get('href'))
    return links
```

```python
def info_wiki(url):
    for link in get_all_links_wiki(url):
        if 'https' in str(link):
            resp = requests.get(link)
        else:
            resp = requests.get('https://ru.wikipedia.org'+str(link))
        ssoup = BeautifulSoup(resp.text, 'html.parser')
        print(len(ssoup.get_text()), ssoup.title)
```

```python
info_wiki('https://ru.wikipedia.org/wiki/Всемирная_библиотека_(Норвежский_книжный_клуб)')
```

# Задача 2

```python
def check_title(ref, n): #Для проверки статьи википедии на то, является ли она служебной
    if 'http' in ref:
        f = requests.get(ref[n])
    else:
        f = requests.get('http://ru.wikipedia.org' + str(ref[n]))  
    soup3 = BeautifulSoup(f.text, 'html.parser')
    tt = str(soup3.title)
    res = ('Википедия:' in tt)
    return res
```

```python
def check_link(link, site): #Проверка полученной ссылки (её формат) и преобразование для дальнейшего использования
    if 'http' in link:
        return str(link)
    else:
        return str('http://'+ site + link)    
```

```python
def first_ref(url): #Поиск первой ссылки на странице
    i = 0
    while check_title(get_all_links_wiki(url), i):
        i += 1
    links = get_all_links_wiki(url)
    return links[i]
```

```python
n = int(input())
```

```python
def second_task(url, n):
    for i in range(n):
        url = check_link(first_ref(url), 'ru.wikipedia.org')
        html = requests.get(url)
        soup = BeautifulSoup(html.text, 'html.parser')
        print(soup.title)
```

```python
second_task('https://ru.wikipedia.org/wiki/Андерсен,_Ханс_Кристиан', n)
```
